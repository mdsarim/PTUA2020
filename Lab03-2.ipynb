## Part 1

```python
import random
import math

n = 0 
d = 0 

conv = float(input("Enter the value of covergence criterion"))
sentinel = int(input("Enter the setinel value"))
iteration = 0
while iteration < sentinel:
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= 1.0:
        n += 1
    d += 1
    iteration +=1
    pi_approx = 4 * n * 1./d
    print(pi_approx)
    if abs(pi_approx-math.pi) / math.pi <= conv:
        print("Draws needed: ", d)
        break
print("Couldn't find a close enough value for pi")
```

## Part 2

```python
import random
import math
import statistics

n = 0 
d = 0 

conv = [0.01, 0.001, 0.0001, 0.00001]

for c in conv:
    steps = []
    for i in range(10):
        while True:
            x = random.random()
            y = random.random()
            if x**2 + y**2 <= 1.0:
                n += 1
            d += 1
            pi_approx = 4 * n * 1./d
            if abs(pi_approx-math.pi) / math.pi <= c:
                break
        steps.append(d)
    avg = statistics.mean(steps)
    sd = statistics.stdev(steps)
    print("With Convergence criterion as " + str(c) + ", Mean number of steps is " + str(avg) + " and the Standard Devation is " + str(sd))
        
```
